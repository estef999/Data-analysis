```python
df_result = df.dropna(thresh=3) # mantine las filas con al menos 3 valores no nulos
print("Después de aplicar dropna(thresh=3):")
print(df_result)
```

```python
df.fillna(df.mean())
df.fillna(df.min())
df.fillna(method="ffill") # foward anterior
df.fillna(method="bfill") # posterior anterior
```

```python
# melt and pivot
data={"nombre":["maria","pedro","juan","pepe"],"mate":[30,30,40,60],"fisica":[40,40,30,39]}
import pandas as pd
df=pd.DataFrame(data)
```

```python
# convertir filsd a columnas  a melt
df_melted=pd.melt(df,id_vars="nombre",var_name="materia",value_name="notas")
```

```python
#pivot table filas a columnas
df_melted.pivot(index="nombre", columns="materia",values="notas")
```
